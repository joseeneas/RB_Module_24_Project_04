```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from unidecode import unidecode
```

```python
race_results18_df = pd.read_csv("../Resources/2013-2023/RaceResults_2018.csv")
race_results19_df = pd.read_csv("../Resources/2013-2023/RaceResults_2019.csv")
race_results20_df = pd.read_csv("../Resources/2013-2023/RaceResults_2020.csv")
race_results21_df = pd.read_csv(("../Resources/2013-2023/RaceResults_2021.csv"),encoding='latin1')
race_results22_df = pd.read_csv(("../Resources/2013-2023/RaceResults_2022.csv"),encoding='latin1')
last_five_years_results_df = pd.read_csv("../Resources/2013-2023/last_five_years_results.csv",encoding='latin1')
```

```python
last_five_years_results_df= last_five_years_results_df.drop(columns=["Fastest Lap"])
```

```python
last_five_years_results_df.head(30)
```

```python
def remove_accents(text):
    return unidecode(text)
last_five_years_results_df['Driver'] = last_five_years_results_df['Driver'].apply(unidecode)
```

```python
# Filter out non-integer values in the "Position" column
last_five_years_results_df = last_five_years_results_df[pd.to_numeric(last_five_years_results_df["Position"], errors='coerce').notnull()]

# Apply the condition
last_five_years_results_df = last_five_years_results_df[last_five_years_results_df["Position"].astype(int) <= 10]
```

```python
#last_five_years_results_df.head(15)
last_five_years_results_df = last_five_years_results_df[last_five_years_results_df["Position"].astype(int) <=10]
```

```python
last_five_years_results_df.head(30)
```

```python
last_five_years_results_df["Team"] = last_five_years_results_df["Team"].replace('Renault','Alpine')
last_five_years_results_df["Team"] = last_five_years_results_df["Team"].replace('Sauber Ferrari','Alfa Romeo Racing Ferrari')
last_five_years_results_df["Team"] = last_five_years_results_df["Team"].replace('Scuderia Toro Rosso Honda','AlphaTauri Honda')


last_five_years_results_df.head(20)
```
