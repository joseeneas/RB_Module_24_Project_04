```python
import pandas as pd
import numpy as np
```

```python
#first we must merge all datasets for 2013-2022
#begin by reading in all the CSV files
results_2013 = pd.read_csv('../Resources/2013-2023/RaceResults_2013.csv')
results_2014 = pd.read_csv('../Resources/2013-2023/RaceResults_2014.csv')
results_2015 = pd.read_csv('../Resources/2013-2023/RaceResults_2015.csv')
results_2016 = pd.read_csv('../Resources/2013-2023/RaceResults_2016.csv')
results_2017 = pd.read_csv('../Resources/2013-2023/RaceResults_2017.csv')
results_2018 = pd.read_csv('../Resources/2013-2023/RaceResults_2018.csv')
results_2019 = pd.read_csv('../Resources/2013-2023/RaceResults_2019.csv')
results_2020 = pd.read_csv('../Resources/2013-2023/RaceResults_2020.csv')
results_2021 = pd.read_csv('../Resources/2013-2023/RaceResults_2021.csv')
results_2022 = pd.read_csv('../Resources/2013-2023/RaceResults_2022.csv')

```

```python
#normalize the data
#2013-2017 datasets do not include the year of the race
results_2013 = results_2013.assign(Year = 2013)
results_2014 = results_2014.assign(Year = 2014)
results_2015 = results_2015.assign(Year = 2015)
results_2016 = results_2016.assign(Year = 2016)
results_2017 = results_2017.assign(Year = 2017)
#results_2017.head(5)

#2021 and 2022 have an extra column to be dropped
results_2021= results_2021.drop(columns=["+1 Pt"])
results_2022= results_2022.drop(columns=["+1 Pt"])
#results_2022.head(5)

#2021 year column has 2020 
results_2021 = results_2021.assign(Year = 2021)
#results_2021.tail(5)

#Time Column is not named the same in all CSV files - 2020 & 2019 are different
results_2019 = results_2019.rename(columns={'Total Time/Gap/Retirement' : "Time/Retired"})
results_2020 = results_2020.rename(columns={'Total Time/Gap/Retirement' : "Time/Retired"})
results_2019
```

```python
#merge into one dataframe
data_2013_2022 = results_2013.merge(results_2014, how='outer')
data_2013_2022 = data_2013_2022.merge(results_2015, how='outer')
data_2013_2022 = data_2013_2022.merge(results_2016, how='outer')
data_2013_2022 = data_2013_2022.merge(results_2017, how='outer')
data_2013_2022 = data_2013_2022.merge(results_2018, how='outer')
data_2013_2022 = data_2013_2022.merge(results_2019, how='outer')
data_2013_2022 = data_2013_2022.merge(results_2020, how='outer')
data_2013_2022 = data_2013_2022.merge(results_2021, how='outer')
data_2013_2022 = data_2013_2022.merge(results_2022, how='outer')
#data_2013_2022.tail(20)
```

```python
data_2013_2022 = data_2013_2022.drop(columns=["No", "Time/Retired", "Points"])
data_2013_2022 = data_2013_2022.rename(columns={'Fastest Lap' : "Fastest Lap Time", "Team" : "Constructor", "Starting Grid" : "Starting Position", \
                                                "Position" : "Ending Position"})
data_2013_2022
```

```python
#reorder the columns
data_2013_2022 = data_2013_2022.reindex(['Year', 'Track', 'Constructor','Driver', 'Starting Position', 'Ending Position', 'Laps', \
                                         'Fastest Lap Time'], axis=1)
data_2013_2022
```

```python
#now we begin cleaning the 1950-2013 dataset

#read in the CSV file
data_1950_2012  = pd.read_csv("../Resources/1950-2013/results.csv")
data_1950_2012
```

```python
#drop columns we won't be using
data_1950_2012 = data_1950_2012.drop(columns=["number", "grid", "positionText", "points", "time", "milliseconds", "fastestLap", "fastestLapSpeed", "rank", "statusId"])
data_1950_2012
```

```python
#renaming one of the columns
data_1950_2012 = data_1950_2012.rename(columns={'fastestLapTime' : "Fastest Lap Time"})
data_1950_2012 = data_1950_2012.rename(columns={'positionOrder' : "Starting Grid"})
data_1950_2012
                                    
```

```python
#we will replace null values with appropriate values
#first the position column
data_1950_2012['Fastest Lap Time'].fillna("DNF", inplace=True)
data_1950_2012['position'].fillna("NC", inplace=True)
```

```python
#now we will start joing with other CSVs to replace the id placeholders
#replace the driver names first
driver_df = pd.read_csv("../Resources/1950-2013/drivers.csv", encoding='latin1')
data_1950_2012 = data_1950_2012.merge(driver_df[["driverId", "forename","surname"]], on="driverId", how='left')
data_1950_2012["Driver"] = data_1950_2012['forename'].astype(str) + " " + data_1950_2012['surname'].astype(str)
data_1950_2012 = data_1950_2012.drop(columns=["driverId", "forename", "surname"])

#replace the raceId next
races_df = pd.read_csv("../Resources/1950-2013/races.csv")
data_1950_2012 = data_1950_2012.merge(races_df[["raceId", "year", "circuitId"]], on="raceId", how = 'left')

#replacing circuitId to get the country
circuits_df = pd.read_csv("../Resources/1950-2013/circuits.csv", encoding='latin1')
data_1950_2012 = data_1950_2012.merge(circuits_df[["circuitId", "country"]], on='circuitId', how = 'left')

#replacing constructorId
constructors_df = pd.read_csv("../Resources/1950-2013/constructors.csv", encoding='latin1')
data_1950_2012 = data_1950_2012.merge(constructors_df[["constructorId", "name"]], on="constructorId", how = "left")

data_1950_2012 = data_1950_2012.drop(columns=["resultId", "raceId", "circuitId", "constructorId"])
data_1950_2012
```

```python
#now we will get rid of data from 2013 to present since we already have it in the other dataframe
data_1950_2012 = data_1950_2012.drop(data_1950_2012[data_1950_2012.year > 2012].index)
```

```python
data_1950_2012 = data_1950_2012.rename(columns={'year' : "Year", "position" : "Ending Position", "Starting Grid" : "Starting Position", "laps" : "Laps", \
                                                "country" : "Track", "name" : "Constructor"})
data_1950_2012
```

```python
data_1950_2012 = data_1950_2012.reindex(['Year', 'Track', 'Constructor','Driver', 'Starting Position', 'Ending Position', 'Laps', \
                                         'Fastest Lap Time'], axis=1)
data_1950_2012.sort_values(by = "Year")
```
