```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
drivers_df = pd.read_csv('LOAD-TEMPLATE-DRIVERS.csv')
results_df = pd.read_csv('LOAD-TEMPLATE-RESULTS.csv')
teams_df = pd.read_csv('LOAD-TEMPLATE-TEAMS.csv')
```


1 breakdown of numbers of races per country 

2 breakdown of numbers of races per track

```python
races_per_country = results_df['Circuit'].value_counts()
# Create a bar chart
plt.figure(figsize=(12, 6))  # Set the figure size
races_per_country.plot(kind='bar', color='skyblue')

# Customize the plot
plt.title('Number of Races per Country(Track)')
plt.xlabel('Country')
plt.ylabel('Number of Races')
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
# Show the plot
plt.tight_layout()
plt.show()
```

# What is the average age of drivers in each country?

```python
# Convert the 'DriverDOB' column to datetime
drivers_df['DriverDOB'] = pd.to_datetime(drivers_df['DriverDOB'])

# Calculate the current year as a reference
current_year = datetime.now().year

# Calculate the age of each driver
drivers_df['Age'] = current_year - drivers_df['DriverDOB'].dt.year

# Calculate the average age of drivers in each country
average_age_by_country = drivers_df.groupby('DriverNationality')['Age'].mean()

# Sort the average age by country in descending order
average_age_by_country = average_age_by_country.sort_values(ascending=False)

# Create a bar plot
plt.figure(figsize=(12, 6))
average_age_by_country.plot(kind='bar', color='skyblue')

# Set plot labels and title
plt.xlabel('Driver Nationality')
plt.ylabel('Average Age')
plt.title('Average Age of Drivers by Nationality')

# Rotate x-axis labels for better readability
plt.xticks(rotation=90)

# Show the plot
plt.tight_layout()
plt.show()
```

# Breakdown of drivers nationalities

```python
nationality_age_df = pd.DataFrame({'Driver Nationality': average_age_by_country.index, 'Average Age': average_age_by_country.values})

# Sort the DataFrame by average age (optional)
nationality_age_df = nationality_age_df.sort_values(by='Average Age', ascending=False)

# Create a bar plot
plt.figure(figsize=(12, 6))
plt.bar(nationality_age_df['Driver Nationality'], nationality_age_df['Average Age'], color='skyblue')

# Set plot labels and title
plt.xlabel('Driver Nationality')
plt.ylabel('Average Age')
plt.title('Average Age of Drivers by Nationality')

# Rotate x-axis labels for better readability
plt.xticks(rotation=90)

# Show the plot
plt.tight_layout()
plt.show()
```
